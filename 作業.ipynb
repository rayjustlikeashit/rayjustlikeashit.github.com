```python
!wget -O taipei_sans_tc_beta.ttf https://drive.google.com/uc?id=1eGAsTN1HBpJAkeVM57_C7ccp7hbgSz3_&export=download #檔案所在的位置

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import io, csv, ssl, urllib.request, requests
from matplotlib.font_manager import FontProperties

matplotlib.font_manager.fontManager.addfont('taipei_sans_tc_beta.ttf')
matplotlib.rc('font', family='Taipei Sans TC Beta') #中文化差件

url = 'https://gis.taiwan.net.tw/od/01_PRD/歷年國內主要觀光遊憩據點遊客人數月別統計.csv'
s=requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
year = int(input('輸入年分'))
city = str(input('縣市別'))
num = int(input('要輸出多少個'))
rows = (df['年別'] == year) & (df['縣市別'] == city)

columns = ['細分', '3月', '6月', '9月', '12月']
result = df.loc[rows, columns].head(num)
result.set_index('細分', inplace=True)

chart = result.plot(kind = 'bar',
           title='各景點旅客人數',  #圖表標題
           xlabel='景點',  #x軸說明文字
           ylabel='人數',  #y軸說明文字
           legend=True,  # 是否顯示圖例
           figsize=(10, 5))  # 圖表大小
                    
plt.show()
```
